```python
import numpy as np
import pandas as pd
```

## pre-processing the data

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits =pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies.head(1)
```

```python
# genres
# id
# keywords
# title
# overview
# cast
movies = movies [["movie_id", "title", "overview", "genres", "keywords", "cast", "crew" ]]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
    lst = []
    for i in ast.literal_eval(obj):
        lst.append(i["name"])
    return lst   
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convert5(obj):
    lst = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=5:
            lst.append(i["name"])
            counter+=1
        else:
            break
    return lst   
```

```python
movies["cast"] = movies["cast"].apply(convert5)
```

```python
movies.head()
```

```python
movies = movies [["movie_id", "title", "overview", "genres", "keywords", "cast"]]
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] +  movies["cast"]
```

```python
movies.head()
```

```python
new_df = movies[["movie_id", "title", "tags"]]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
## vectorizing the words to give better recommendations
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vactors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vactors[0]
```

```python
(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vactors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:7]
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:7]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
      
```

```python
recommend("Batman Begins")
```
